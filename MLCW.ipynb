<p> HST:	Total	number	of	shots	on	target	by	the	home	team</p>

<p> HELO:	Home team ELO score</p>
<p> AELO:	Away team ELO score</p>
<p> FTR: Full time results</p>

```python
AST = []
HST = []
HF = []
AF = []
HC = []
AC = []
HELO = []
AELO = []

for index, row in dataset_2018.iterrows():
    sumAST = 0
    sumHST = 0
    sumHF = 0
    sumAF = 0
    sumHC = 0
    sumAC = 0
    numHELO = 0
    numAELO = 0
    counter = 0
    
    
    for j, data in dataset.iterrows():
        if row['HomeTeam'] == data['HomeTeam'] and row['AwayTeam'] == data['AwayTeam']:
            sumAST += data['AST']
            sumHST += data['HST']
            sumHF += data['HF']
            sumAF += data['AF']
            sumHC += data['HC']
            sumAC += data['AC']
            numHELO = data['HELO']
            numAELO = data['AELO']
            counter += 1
    
    HST.append(round(sumHST / counter))
    AST.append(round(sumAST / counter))
    HF.append(round(sumHF / counter))
    AF.append(round(sumAF / counter))
    HC.append(round(sumHC / counter))
    AC.append(round(sumAC / counter))
    HELO.append(numHELO)
    AELO.append(numAELO)
```

```python
dataset_2018['HST'] = HST
dataset_2018['AST'] = AST
dataset_2018['HF'] = HF
dataset_2018['AF'] = AF
dataset_2018['HC'] = HC
dataset_2018['AC'] = AC
dataset_2018['HELO'] = HELO
dataset_2018['AELO'] = AELO
```

<h3>4. Methodology Overview</h3>

<p>From the give dataset I chose 3 features (shots on target, corners and faults) to which I have added each team's current EOL rating (http://clubelo.com/ENG). Then, I chose 5 main popular models to train my features on and compared their resulted accuracy, based on which I have chosed the final model.</p>

<h3>5. Model training/validation</h3>

```python
X = features[['HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HELO', 'AELO']]
y = features['FTR']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
classifier = DecisionTreeClassifier()
classifier2 = GaussianNB()
classifier3 = KNeighborsClassifier()
classifier4 = LogisticRegression()
classifier5 = RandomForestClassifier()
```

```python
y_pred = classifier.fit(X_train, y_train).predict(X_test)
y_pred2 = classifier2.fit(X_train, y_train).predict(X_test)
y_pred3 = classifier3.fit(X_train, y_train).predict(X_test)
y_pred4 = classifier4.fit(X_train, y_train).predict(X_test)
y_pred5 = classifier5.fit(X_train, y_train).predict(X_test)
```

<h3>6. Results</h3>

```python
accuracy_score(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred2)
```

```python
accuracy_score(y_test, y_pred3)
```

```python
accuracy_score(y_test, y_pred4)
```

```python
accuracy_score(y_test, y_pred5)
```

<p>After testing each calssifier's prediction accuracy, it is determined that the fourth, LogisticRegression has the highest accuracy (59%). Before adding the ELO, GaussianNB was leading with 53%</p>
